```python
import pandas as pd
```

```python
data = pd.read_csv('../data/preprocessed/SanDiego_from_1980_preprocessed.csv', parse_dates=['YEARMODA'])
data.head()
```

```python
data.columns
```

```python
subset = data.set_index('YEARMODA')
subset = subset[['STN', 'TEMP', 'MAX', 'MIN']]
subset = subset.dropna()
```

```python
sd_airport = subset[subset['STN'] == 722900] # SAN DIEGO INTERNATIONAL AIRPO
sd_airport.head()
```

```python
from plotly import plotly as ply
from plotly import graph_objs as go
```

```python
sd_airport_oneday = sd_airport[(sd_airport.index.month == 12) & (sd_airport.index.day == 2)]
sd_airport_2014 = sd_airport['2014']
```

```python
trace = go.Scatter(
    x=sd_airport_2014.index,
    y=sd_airport_2014['TEMP'],
)
layout = go.Layout(
    title="San Diego Weather 2014"
)
figure = go.Figure(data=[trace], layout=layout)
ply.iplot(figure, filename='WeatherData/sandiego')
```

```python
trace0 = go.Scatter(
    x=sd_airport_2014.index,
    y=sd_airport_2014['TEMP'],
    name='Mean'
)
trace1 = go.Scatter(
    x=sd_airport_2014.index,
    y=sd_airport_2014['MAX'],
    name='Max'
)
trace2 = go.Scatter(
    x=sd_airport_2014.index,
    y=sd_airport_2014['MIN'],
    name='Min'
)
layout = go.Layout(
    title="San Diego Weather 2014 Min & Max"
)
figure = go.Figure(data=[trace1, trace0, trace2], layout=layout)
ply.iplot(figure, filename='WeatherData/sandiego')
```

```python
ply.image.save_as(figure, '../figures/San Diego Weather 2014.png')
```

```python
sd_airport['day'] = sd_airport.index.dayofyear
sd_airport['year'] = sd_airport.index.year
```

```python
sd_airport2 = sd_airport['1981-1-1':'2014-12-31']
```

```python
day_year = sd_airport2.pivot(index='day', columns='year', values='TEMP')
day_year = day_year.fillna(0)
day_year
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
projected = pca.fit_transform(day_year.T)
```

```python
projected.shape
```

```python
go.Scatter()
```

```python
trace0 = go.Scatter(
    x=projected[:,0],
    y=projected[:,1],
    name='Projected',
    mode='text',
    text=day_year.columns,
)
layout = go.Layout(
    title="San Diego Temperature PCA"
)
figure = go.Figure(data=[trace0], layout=layout)
ply.iplot(figure, filename='WeatherData/sandiego_PCA')
```

```python
ply.image.save_as(figure, '../figures/San Diego Weather PCA.png')
```
